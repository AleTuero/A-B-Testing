```python
import pandas as pd
import numpy as np

# =========================
# CONFIG
# =========================
CSV_PATH = "df_client_usage_head10.csv"   # <- cambia si tu archivo se llama distinto
TOP_N = 10

# =========================
# LOAD
# =========================
df_top10 = pd.read_csv(CSV_PATH)

print("✅ Cargado:", CSV_PATH)
print("Shape:", df_top10.shape)
display(df_top10.head())

# =========================
# HELPERS: detectar columnas de uso (calls/logons)
# =========================
cols_lower = {c: c.lower() for c in df_top10.columns}
call_cols = [c for c in df_top10.columns if "call" in cols_lower[c]]
logon_cols = [c for c in df_top10.columns if "logon" in cols_lower[c] or "login" in cols_lower[c]]

# =========================
# CLEAN TYPES
# =========================
# intentamos convertir a numérico las columnas típicas si existen
num_candidates = ["clnt_tenure_yr", "clnt_tenure_mnth", "clnt_age", "num_accts", "bal"] + call_cols + logon_cols
for c in num_candidates:
    if c in df_top10.columns:
        df_top10[c] = pd.to_numeric(df_top10[c], errors="coerce")

# =========================
# OPTIONAL: usage_score (si existen calls y logons)
# =========================
if len(call_cols) > 0 or len(logon_cols) > 0:
    df_top10["usage_score"] = 0
    if len(call_cols) > 0:
        df_top10["usage_score"] += df_top10[call_cols].fillna(0).sum(axis=1)
    if len(logon_cols) > 0:
        df_top10["usage_score"] += df_top10[logon_cols].fillna(0).sum(axis=1)

# =========================
# SEGMENTATION BINS
# =========================
if "clnt_age" in df_top10.columns:
    df_top10["age_group"] = pd.cut(
        df_top10["clnt_age"],
        bins=[0, 30, 40, 50, 60, 120],
        labels=["<=30", "31-40", "41-50", "51-60", "60+"],
        right=True
    )

if "clnt_tenure_yr" in df_top10.columns:
    df_top10["tenure_group"] = pd.cut(
        df_top10["clnt_tenure_yr"],
        bins=[-0.1, 1, 3, 5, 10, 100],
        labels=["<=1y", "1-3y", "3-5y", "5-10y", "10y+"],
        right=True
    )

# =========================
# INSIGHTS: TABLES
# =========================
print("\n=========================")
print("1) DESCRIPTIVOS (NUMÉRICAS)")
print("=========================")

num_cols = [c for c in ["clnt_age", "clnt_tenure_yr", "clnt_tenure_mnth", "num_accts", "bal", "usage_score"] if c in df_top10.columns]
if num_cols:
    desc = df_top10[num_cols].describe().T
    display(desc)
else:
    print("No se encontraron columnas numéricas estándar para describir.")

print("\n=========================")
print("2) DISTRIBUCIÓN (EDAD / TENURE / GÉNERO)")
print("=========================")

if "age_group" in df_top10.columns:
    age_dist = df_top10["age_group"].value_counts(dropna=False)
    age_pct = df_top10["age_group"].value_counts(normalize=True, dropna=False).mul(100).round(1)
    display(pd.DataFrame({"count": age_dist, "pct": age_pct}))
else:
    print("No existe 'clnt_age' para crear age_group.")

if "tenure_group" in df_top10.columns:
    ten_dist = df_top10["tenure_group"].value_counts(dropna=False)
    ten_pct = df_top10["tenure_group"].value_counts(normalize=True, dropna=False).mul(100).round(1)
    display(pd.DataFrame({"count": ten_dist, "pct": ten_pct}))
else:
    print("No existe 'clnt_tenure_yr' para crear tenure_group.")

if "gendr" in df_top10.columns:
    g_dist = df_top10["gendr"].value_counts(dropna=False)
    g_pct = df_top10["gendr"].value_counts(normalize=True, dropna=False).mul(100).round(1)
    display(pd.DataFrame({"count": g_dist, "pct": g_pct}))
else:
    print("No existe columna 'gendr'.")

print("\n=========================")
print("3) USO (CALLS / LOGONS)")
print("=========================")

usage_cols = call_cols + logon_cols
if usage_cols:
    usage_desc = df_top10[usage_cols].describe().T
    display(usage_desc)
else:
    print("No se detectaron columnas de uso (call/logon/login).")

print("\n=========================")
print("4) CRUCE EDAD x TENURE (porcentaje dentro de los 10)")
print("=========================")

if "age_group" in df_top10.columns and "tenure_group" in df_top10.columns:
    cross = pd.crosstab(df_top10["age_group"], df_top10["tenure_group"], normalize="all").mul(100).round(1)
    display(cross)
else:
    print("Falta age_group o tenure_group para el cruce.")

# =========================
# INSIGHTS: AUTO-TEXT (resumen interpretado)
# =========================
print("\n=========================")
print("5) RESUMEN AUTOMÁTICO (texto listo)")
print("=========================")

lines = []

# Edad
if "clnt_age" in df_top10.columns:
    age_med = df_top10["clnt_age"].median()
    age_mean = df_top10["clnt_age"].mean()
    lines.append(f"- Edad: media ≈ {age_mean:.1f} años; mediana = {age_med:.1f}.")
    if "age_group" in df_top10.columns:
        top_age_group = df_top10["age_group"].value_counts().idxmax()
        lines.append(f"  Grupo de edad más frecuente: {top_age_group}.")
else:
    lines.append("- Edad: no disponible (no hay 'clnt_age').")

# Tenure
if "clnt_tenure_yr" in df_top10.columns:
    ten_med = df_top10["clnt_tenure_yr"].median()
    ten_mean = df_top10["clnt_tenure_yr"].mean()
    lines.append(f"- Antigüedad: media ≈ {ten_mean:.1f} años; mediana = {ten_med:.1f}.")
    if "tenure_group" in df_top10.columns:
        top_ten_group = df_top10["tenure_group"].value_counts().idxmax()
        lines.append(f"  Tramo de antigüedad más frecuente: {top_ten_group}.")
else:
    lines.append("- Antigüedad: no disponible (no hay 'clnt_tenure_yr').")

# Finanzas
if "bal" in df_top10.columns:
    bal_med = df_top10["bal"].median()
    bal_mean = df_top10["bal"].mean()
    lines.append(f"- Balance: media ≈ {bal_mean:,.2f}; mediana = {bal_med:,.2f}.")
else:
    lines.append("- Balance: no disponible (no hay 'bal').")

# Complejidad
if "num_accts" in df_top10.columns:
    lines.append(f"- Nº de cuentas: media ≈ {df_top10['num_accts'].mean():.1f}; mediana = {df_top10['num_accts'].median():.1f}.")
else:
    lines.append("- Nº de cuentas: no disponible (no hay 'num_accts').")

# Uso
if "usage_score" in df_top10.columns:
    lines.append(f"- Uso (usage_score): media ≈ {df_top10['usage_score'].mean():.1f}; mediana = {df_top10['usage_score'].median():.1f}.")
    if call_cols:
        lines.append(f"  Calls columns detectadas: {call_cols}")
    if logon_cols:
        lines.append(f"  Logon columns detectadas: {logon_cols}")
else:
    lines.append("- Uso: no calculado (no se detectaron columnas call/logon).")

# Nota de tamaño muestral
lines.append(f"- Nota: estos insights son exploratorios (muestra N={len(df_top10)}).")

print("\n".join(lines))
```

```python
import pandas as pd
import numpy as np

# =========================
# CONFIG
# =========================
INPUT_PATH = "df_final_web_data.csv"          # <-- tu CSV
OUTPUT_EXCEL = "web_funnel_clean_one_sheet.xlsx"

VALID_STEPS = ["start", "step_1", "step_2", "step_3", "confirm"]
STEP_ORDER = {s: i + 1 for i, s in enumerate(VALID_STEPS)}

# =========================
# 1) Load + tipos
# =========================
df = pd.read_csv(INPUT_PATH)

# Normaliza strings
for c in ["visit_id", "process_step"]:
    if c in df.columns:
        df[c] = df[c].astype(str).str.strip()

# Parse datetime
if "date_time" not in df.columns:
    raise ValueError("No existe la columna 'date_time' en el dataset.")
df["date_time"] = pd.to_datetime(df["date_time"], errors="coerce")

# =========================
# 2) Filtros básicos (ruido obvio)
# =========================
df = df.dropna(subset=["date_time"])
df = df[df["visit_id"].notna() & (df["visit_id"].astype(str).str.len() > 0)]
df = df[df["process_step"].isin(VALID_STEPS)].copy()

df["step_order"] = df["process_step"].map(STEP_ORDER).astype(int)

# =========================
# 3) DEDUP (3 capas)
# =========================
# 3.1 exact duplicates (todas las columnas)
df = df.drop_duplicates()

# 3.2 duplicados dentro de sesión a mismo timestamp y step
df = df.drop_duplicates(subset=["visit_id", "date_time", "process_step"])

# 3.3 loops: mismo step repetido en una sesión -> nos quedamos con la PRIMERA vez
df = df.sort_values(["visit_id", "date_time", "step_order"])
df = df.drop_duplicates(subset=["visit_id", "process_step"], keep="first").copy()

# =========================
# 4) Mantener trazabilidad (orden funnel correcto)
#     step_order estrictamente creciente dentro de cada visit_id
# =========================
df["prev_step_order"] = df.groupby("visit_id")["step_order"].shift(1)
df["is_break"] = (df["prev_step_order"].notna()) & (df["step_order"] <= df["prev_step_order"])
df["bad_session"] = df.groupby("visit_id")["is_break"].transform("any")

df_clean = df[~df["bad_session"]].copy()
df_clean = df_clean.drop(columns=["prev_step_order", "is_break", "bad_session"])

# =========================
# 5) TIEMPOS ENTRE STEPS (fila a fila)
# =========================
df_clean = df_clean.sort_values(["visit_id", "date_time", "step_order"]).copy()

df_clean["prev_date_time"] = df_clean.groupby("visit_id")["date_time"].shift(1)
df_clean["time_from_prev_step_sec"] = (df_clean["date_time"] - df_clean["prev_date_time"]).dt.total_seconds()

df_clean["start_time"] = df_clean.groupby("visit_id")["date_time"].transform("min")
df_clean["time_from_start_sec"] = (df_clean["date_time"] - df_clean["start_time"]).dt.total_seconds()

# Guardrails (por si acaso)
df_clean.loc[df_clean["time_from_prev_step_sec"] < 0, "time_from_prev_step_sec"] = np.nan
df_clean.loc[df_clean["time_from_start_sec"] < 0, "time_from_start_sec"] = np.nan

# =========================
# 6) TIEMPOS POR TRAMO (resumen por sesión) + MERGE a df_clean
# =========================
step_times = (
    df_clean.pivot_table(
        index="visit_id",
        columns="process_step",
        values="date_time",
        aggfunc="min"
    )
)

session_deltas = pd.DataFrame(index=step_times.index)
for a, b in zip(VALID_STEPS[:-1], VALID_STEPS[1:]):
    session_deltas[f"sec_{a}to{b}"] = (step_times[b] - step_times[a]).dt.total_seconds()

session_deltas["sec_start_to_confirm"] = (step_times["confirm"] - step_times["start"]).dt.total_seconds()

# Guardrails no negativos
for c in session_deltas.columns:
    session_deltas.loc[session_deltas[c] < 0, c] = np.nan

session_deltas = session_deltas.reset_index()

# Merge: añade columnas de sesión a cada fila del funnel
final_df = df_clean.merge(session_deltas, on="visit_id", how="left")

# =========================
# 7) Export EXCEL (UNA SOLA HOJA)
# =========================
with pd.ExcelWriter(OUTPUT_EXCEL, engine="openpyxl") as writer:
    final_df.to_excel(writer, sheet_name="clean_funnel", index=False)

print("✅ Listo")
print(" - Filas finales:", len(final_df))
print(" - Sesiones únicas:", final_df["visit_id"].nunique())
print("💾 Excel:", OUTPUT_EXCEL)
```

```python
#Métricas de rendimiento
#Tasa finalización del proceso por variación (start -> confirm)
df = pd.read_excel("web_funnel_with_variation.xlsx")
# Crear una columna que indique si la sesión llegó al paso 'confirm'
df['completed'] = df['process_step'] == 'confirm'
# Agrupar por 'Variation' y calcular las métricas
metrics = df.groupby('Variation').agg(
    total_sessions=('visit_id', 'nunique'),
    completed_sessions=('completed', 'sum')
).reset_index()
# Calcular la tasa de finalización
metrics['completion_rate'] = (metrics['completed_sessions'] / metrics['total_sessions']) * 100
# Mostrar los resultados
print(metrics)
```

```python
#Métricas de rendimiento sin variacion
#Tasa finalización del proceso por variación (start -> confirm) sin tener en cuenta Variation
df = pd.read_excel("web_funnel_with_variation.xlsx")
# Crear una columna que indique si la sesión llegó al paso 'confirm'    
df['completed'] = df['process_step'] == 'confirm'
# Agrupar por 'visit_id' para obtener sesiones únicas
session_metrics = df.groupby('visit_id').agg(
    variation=('Variation', 'first'),
    completed=('completed', 'max')
).reset_index()
# Calcular las métricas generales
total_sessions = session_metrics['visit_id'].nunique()
completed_sessions = session_metrics['completed'].sum()
completion_rate = (completed_sessions / total_sessions) * 100
# Mostrar los resultados
print(f"Total sessions: {total_sessions}")
print(f"Completed sessions: {completed_sessions}")
print(f"Completion rate: {completion_rate:.2f}%")
```

```python
# Análisis de retrocesos en el proceso (backtracking errors)
import pandas as pd
import numpy as np
INPUT_PATH = "df_final_web_data.csv"
VALID_STEPS = ["start", "step_1", "step_2", "step_3", "confirm"]
STEP_ORDER = {s: i + 1 for i, s in enumerate(VALID_STEPS)}
# =========================
# 1) Cargar "raw" y filtros mínimos (NO aplanar)
# =========================
raw = pd.read_csv(INPUT_PATH)
for c in ["visit_id", "process_step"]:
    if c in raw.columns:
        raw[c] = raw[c].astype(str).str.strip()
raw["date_time"] = pd.to_datetime(raw["date_time"], errors="coerce")
raw = raw.dropna(subset=["date_time"])
raw = raw[raw["visit_id"].notna() & (raw["visit_id"].astype(str).str.len() > 0)]
raw = raw[raw["process_step"].isin(VALID_STEPS)].copy()
raw["step_order"] = raw["process_step"].map(STEP_ORDER).astype(int)
# (Opcional) dedup SOLO de ruido "mismo timestamp mismo step" (no mata retrocesos)
raw = raw.drop_duplicates(subset=["visit_id", "date_time", "process_step"])

# =========================
# 2) Detectar retrocesos (errores)
# =========================
raw = raw.sort_values(["visit_id", "date_time", "step_order"]).copy()
raw["prev_step"] = raw.groupby("visit_id")["process_step"].shift(1)
raw["prev_order"] = raw.groupby("visit_id")["step_order"].shift(1)
# Error si el orden baja respecto al anterior
raw["is_backtrack_error"] = raw["prev_order"].notna() & (raw["step_order"] < raw["prev_order"])
# Para análisis: transición que causó el error
raw["transition"] = np.where(
    raw["prev_step"].notna(),
    raw["prev_step"].astype(str) + " -> " + raw["process_step"].astype(str),
    np.nan
)

# =========================
# 3) KPIs principales
# =========================
# Total de transiciones (cada fila excepto la primera de cada sesión)
raw["is_transition"] = raw["prev_order"].notna()
total_transitions = int(raw["is_transition"].sum())
total_errors = int(raw["is_backtrack_error"].sum())
kpi_error_rate = total_errors / total_transitions if total_transitions else np.nan
# % de sesiones con al menos un retroceso
session_has_error = raw.groupby("visit_id")["is_backtrack_error"].any()
kpi_sessions_with_error = session_has_error.mean()  # proporción
print("=== KPI Retrocesos ===")
print("Transiciones totales:", total_transitions)
print("Errores (retrocesos) :", total_errors)
print("Error rate (errores/transiciones):", round(kpi_error_rate, 4) if pd.notna(kpi_error_rate) else np.nan)
print("% sesiones con >=1 error:", round(float(kpi_sessions_with_error), 4))

# =========================
# 4) Breakdown útil
# =========================
# Errores por transición exacta (ej: step_2 -> step_1)
errors_by_transition = (
    raw.loc[raw["is_backtrack_error"], "transition"]
       .value_counts()
       .rename_axis("transition")
       .reset_index(name="n_errors")
)
# Errores "originados" desde qué step (prev_step)
errors_by_prev_step = (
    raw.loc[raw["is_backtrack_error"], "prev_step"]
       .value_counts()
       .rename_axis("prev_step")
       .reset_index(name="n_errors")
)
print("\nTop transiciones de error:")
print(errors_by_transition.head(10).to_string(index=False))
print("\nErrores por step de origen (desde dónde retroceden):")
print(errors_by_prev_step.to_string(index=False))

import pandas as pd

# ======================================================
# 5) TASA DE ERROR POR PASO (desde qué paso retroceden)
# ======================================================

# Asegurar que no haya formato global raro de floats
pd.reset_option("display.float_format")

# Solo transiciones válidas (excluye primera fila de cada sesión)
trans = raw[raw["is_transition"]].copy()

error_rate_by_prev_step = (
    trans.groupby("prev_step", dropna=False)
         .agg(
             n_transitions=("is_transition", "sum"),
             n_backtracks=("is_backtrack_error", "sum")
         )
         .assign(
             error_rate=lambda x: x["n_backtracks"] / x["n_transitions"],
             error_rate_pct=lambda x: (x["n_backtracks"] / x["n_transitions"]) * 100
         )
         .reset_index()
         .sort_values("error_rate", ascending=False)
)

print("\n=== Error rate por paso (desde prev_step) ===")
print(
    error_rate_by_prev_step.to_string(
        index=False,
        formatters={
            "error_rate": "{:.4f}".format,
            "error_rate_pct": "{:.2f}%".format
        }
    )
)

```

```python
 # Contraste de hipótesis
# H0 (nula): la tasa de finalización del proceso es igual en Test y Control
# H1 (alternativa): la tasa de finalización del proceso es mayor en Test que en Control
from statsmodels.stats.proportion import proportions_ztest
# =========================
# 1) Cargar datos
# =========================
df = pd.read_excel("web_funnel_with_variation.xlsx")
# =========================
# 2) Definir sesión completada
# =========================
df["completed"] = df["process_step"] == "confirm"
# =========================
# 3) Métricas a nivel SESIÓN
# =========================
session_metrics = (
    df.groupby("Variation")
      .agg(
          total_sessions=("visit_id", "nunique"),
          completed_sessions=("completed", "sum")
      )
      .loc[["Test", "Control"]]   # 🔑 orden explícito
)
# =========================
# 4) Preparar datos para el test
# =========================
counts = session_metrics["completed_sessions"].values
nobs   = session_metrics["total_sessions"].values
# =========================
# 5) Test de proporciones
# =========================
z_stat, p_value = proportions_ztest(
    count=counts,
    nobs=nobs,
    alternative="larger"  # Test > Control
)
# =========================
# 6) Resultados
# =========================
print("=== Test de hipótesis: Completion Rate ===")
print(session_metrics)
print(f"\nZ-statistic: {z_stat:.3f}")
print(f"P-value: {p_value:.6f}")
alpha = 0.05
if p_value < alpha:
    print("✅ Rechazamos H0: la variación Test mejora significativamente la tasa de finalización.")
else:
    print("❌ No rechazamos H0: no hay evidencia suficiente.")
```

```python
#Coste efectividad
import pandas as pd
import numpy as np
from math import sqrt
from scipy.stats import norm
# ============
# Load + session-level completion
# ============
df = pd.read_excel("web_funnel_with_variation.xlsx")
df["completed"] = df["process_step"] == "confirm"
session_metrics = (
    df.groupby("Variation")
      .agg(total_sessions=("visit_id", "nunique"),
           completed_sessions=("completed", "sum"))
      .loc[["Test", "Control"]]
)
n_test = int(session_metrics.loc["Test", "total_sessions"])
n_ctrl = int(session_metrics.loc["Control", "total_sessions"])
x_test = int(session_metrics.loc["Test", "completed_sessions"])
x_ctrl = int(session_metrics.loc["Control", "completed_sessions"])
p_test = x_test / n_test
p_ctrl = x_ctrl / n_ctrl
diff_abs = p_test - p_ctrl                 # diferencia absoluta (pp en decimal)
lift_rel = diff_abs / p_ctrl               # mejora relativa
# ============
# 95% CI para diff_abs (unpooled)
# ============
se_unpooled = sqrt(p_test*(1-p_test)/n_test + p_ctrl*(1-p_ctrl)/n_ctrl)
ci_low = diff_abs - 1.96 * se_unpooled
ci_high = diff_abs + 1.96 * se_unpooled
print("=== Cost-effectiveness (Completion) ===")
print(f"p_test={p_test:.4%} | p_ctrl={p_ctrl:.4%}")
print(f"Diff abs={diff_abs:.4%} (pp) | Lift rel={lift_rel:.2%}")
print(f"95% CI(diff abs)=[{ci_low:.4%}, {ci_high:.4%}]")
# ============
# A) Umbral +5 puntos porcentuales (pp)
# ============
threshold_pp = 0.05
meets_pp = diff_abs >= threshold_pp
print("\nA) Umbral +5pp:")
print("Meets threshold?:", meets_pp)
# Test one-sided contra el umbral (aprox normal):
# H0: diff_abs <= 0.05  vs  H1: diff_abs > 0.05
z_thr = (diff_abs - threshold_pp) / se_unpooled
p_thr = 1 - norm.cdf(z_thr)
print(f"Z(threshold)={z_thr:.3f} | p-value(H1: diff>0.05)={p_thr:.6f}")
# ============
# B) Umbral +5% relativo
# ============
threshold_rel = 0.05
meets_rel = lift_rel >= threshold_rel
print("\nB) Umbral +5% relativo:")
print("Meets threshold?:", meets_rel)
# ============
# Conclusión "business"
# ============
if meets_pp or meets_rel:
    print("\n✅ Conclusión negocio: la mejora observada cumple el umbral de coste-efectividad.")
else:
    print("\n❌ Conclusión negocio: la mejora observada NO cumple el umbral de coste-efectividad.")
```

```python
#Hipotesis adicional
# H0: la edad media de los usuarios del Test es igual a la del Control
#H1: la edad media de los usuarios del Test es distinta a la del Control
 
from scipy.stats import ttest_ind
# =========================
# 1) Cargar datasets
# =========================
df_web = pd.read_excel("web_funnel_with_variation.xlsx")
df_master = pd.read_csv("df_master_with_value_category.csv")
# =========================
# 2) Quedarse con una fila por sesión
# =========================
df_sessions = (
    df_web.drop_duplicates(subset=["visit_id"])
          .loc[:, ["visit_id", "client_id", "Variation"]]
)
# =========================
# 3) Join con master (edad)
# =========================
df_sessions = (
    df_sessions
    .merge(
        df_master[["client_id", "clnt_age"]],
        on="client_id",
        how="left"
    )
    .dropna(subset=["Variation", "clnt_age"])
)
# =========================
# 4) Separar grupos
# =========================
age_test = df_sessions[df_sessions["Variation"] == "Test"]["clnt_age"]
age_control = df_sessions[df_sessions["Variation"] == "Control"]["clnt_age"]
# =========================
# 5) Welch t-test
# =========================
t_stat, p_value = ttest_ind(
    age_test,
    age_control,
    equal_var=False,
    nan_policy="omit"
)
# =========================
# 6) Resultados
# =========================
print("=== Hipótesis adicional: Edad media ===")
print(f"Edad media Test: {age_test.mean():.2f}")
print(f"Edad media Control: {age_control.mean():.2f}")
print(f"Diferencia (Test - Control): {(age_test.mean() - age_control.mean()):.2f} años")
print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.6f}")
alpha = 0.05
if p_value < alpha:
    print("✅ Rechazamos H0: existen diferencias significativas en la edad media entre Test y Control.")
else:
    print("❌ No rechazamos H0: no se observan diferencias significativas en la edad media.")
```
